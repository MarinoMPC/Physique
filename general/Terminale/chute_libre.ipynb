```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#### Paramètres ####

# Paramètres communs
g=9.81
tmax=4
xmax=142
m=10

# Système 1 #

alpha=70   # in degrees
v0=40


alpha_rad=alpha*np.pi/180

# Système 2 #
compare=True
alpha_2=30
v0_2=30

alpha_2_rad=alpha_2*np.pi/180
```

```python
## Équations horaires ##
t=np.linspace(0,tmax,1000)
x_t=v0*np.cos(alpha_rad)*t
z_t=-g/2*t**2+v0*np.sin(alpha_rad)*t

x_t_1=v0*np.cos(alpha_2_rad)*t
z_t_1=-g/2*t**2+v0*np.sin(alpha_2_rad)*t
x_t_2=v0_2*np.cos(alpha_rad)*t
z_t_2=-g/2*t**2+v0_2*np.sin(alpha_rad)*t
x_t_3=v0_2*np.cos(alpha_2_rad)*t
z_t_3=-g/2*t**2+v0_2*np.sin(alpha_2_rad)*t


fig,(ax0,ax1)=plt.subplots(1,2,figsize=(16,6))

plt.sca(ax0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax0.plot(t,x_t,label=r'$\alpha$='+str(alpha)+r'°, $v_0$='+str(v0)+'m/s')
ax0.set_xlabel('t (s)',fontsize=17)
ax0.set_ylabel('x (m)',fontsize=17)
ax0.set_xlim(0)
ax0.set_ylim(0)

if compare==True:
    ax0.plot(t,x_t_1,label=r'$\alpha$='+str(alpha_2)+r'°, $v_0$='+str(v0)+'m/s')
    ax0.plot(t,x_t_2,label=r'$\alpha$='+str(alpha)+r'°, $v_0$='+str(v0_2)+'m/s')
    ax0.plot(t,x_t_3,label=r'$\alpha$='+str(alpha_2)+r'°, $v_0$='+str(v0_2)+'m/s')
    plt.legend(fontsize=17,ncol=2)

plt.sca(ax1)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax1.plot(t,z_t,label=r'$\alpha$='+str(alpha)+r'°, $v_0$='+str(v0)+'m/s')
ax1.set_xlabel('t (s)',fontsize=17)
ax1.set_ylabel('z (m)',fontsize=17)
ax1.set_xlim(0)

if compare==True:
    ax1.plot(t,z_t_1,label=r'$\alpha$='+str(alpha_2)+r'°, $v_0$='+str(v0)+'m/s')
    ax1.plot(t,z_t_2,label=r'$\alpha$='+str(alpha)+r'°, $v_0$='+str(v0_2)+'m/s')
    ax1.plot(t,z_t_3,label=r'$\alpha$='+str(alpha_2)+r'°, $v_0$='+str(v0_2)+'m/s')
    plt.legend(fontsize=17)
plt.savefig('equations_horaires')
```

```python
## Équation de la trajectoire ##
x=np.linspace(0,xmax,1000)
z_x=-g/(2*v0**2*np.cos(alpha_rad)**2)*x**2+np.tan(alpha_rad)*x

if compare==True:
    z_x_1=-g/(2*v0**2*np.cos(alpha_2_rad)**2)*x**2+np.tan(alpha_2_rad)*x
    z_x_2=-g/(2*v0_2**2*np.cos(alpha_rad)**2)*x**2+np.tan(alpha_rad)*x
    z_x_3=-g/(2*v0_2**2*np.cos(alpha_2_rad)**2)*x**2+np.tan(alpha_2_rad)*x


fig,ax=plt.subplots(figsize=(7,7))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax.plot(x,z_x,label=r'$\alpha$='+str(alpha)+r'°, $v_0$='+str(v0)+'m/s')
ax.set_xlabel('x (m)',fontsize=17)
ax.set_ylabel('z (m)',fontsize=17)
ax.set_xlim(0,xmax)
ax.set_ylim(0,xmax)

if compare==True:
    ax.plot(x,z_x_1,label=r'$\alpha$='+str(alpha_2)+r'°, $v_0$='+str(v0)+'m/s')
    ax.plot(x,z_x_2,label=r'$\alpha$='+str(alpha)+r'°, $v_0$='+str(v0_2)+'m/s')
    ax.plot(x,z_x_3,label=r'$\alpha$='+str(alpha_2)+r'°, $v_0$='+str(v0_2)+'m/s')


    plt.legend(fontsize=17)
plt.savefig('equation_trajectoire')
```

```python
alpha,alpha_1,alpha_2=30,45,75
alpha_rad,alpha_1_rad,alpha_2_rad=alpha*np.pi/180,alpha_1*np.pi/180,alpha_2*np.pi/180
v0,v0_1,v0_2=20,30,40
```

```python
### Portée ###
alpha_var=np.linspace(0,90,1000)*np.pi/180
xp_alpha=v0**2*np.sin(2*alpha_var)/g
xp_alpha_1=v0_1**2*np.sin(2*alpha_var)/g
xp_alpha_2=v0_2**2*np.sin(2*alpha_var)/g

v0_var=np.linspace(0,100,1000)
xp_v0=v0_var**2*np.sin(2*alpha_rad)/g
xp_v0_1=v0_var**2*np.sin(2*alpha_1_rad)/g
xp_v0_2=v0_var**2*np.sin(2*alpha_2_rad)/g




fig,(ax0,ax1)=plt.subplots(1,2,figsize=(15,6))

plt.sca(ax0)
ax0.plot(alpha_var*180/np.pi,xp_alpha,label=r'$v_0$='+str(v0)+'m/s')
ax0.plot(alpha_var*180/np.pi,xp_alpha_1,label=r'$v_0$='+str(v0_1)+'m/s')
ax0.plot(alpha_var*180/np.pi,xp_alpha_2,label=r'$v_0$='+str(v0_2)+'m/s')

ax0.set_xlabel(r'$\alpha$ (°)',fontsize=17)
ax0.set_ylabel('Portée (m)', fontsize=17)
ax0.set_xlim(0,90)
ax0.set_ylim(0,240)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='upper left',ncol=1,fontsize=17)

plt.sca(ax1)
ax1.plot(v0_var,xp_v0,label=r'$\alpha$='+str(alpha)+'°')
ax1.plot(v0_var,xp_v0_1,label=r'$\alpha$='+str(alpha_1)+'°')
ax1.plot(v0_var,xp_v0_2,label=r'$\alpha$='+str(alpha_2)+'°')
ax1.set_xlabel(r'$v_0$ (m/s)',fontsize=17)
ax1.set_ylabel('Portée (m)', fontsize=17)
ax1.set_xlim(0,100)
ax1.set_ylim(0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=17)
plt.savefig('portee')

```

```python
### Flèche ###

alpha_var=np.linspace(0,90,1000)*np.pi/180
H_alpha=v0**2*np.sin(alpha_var)**2/2/g
H_alpha_1=v0_1**2*np.sin(alpha_var)**2/2/g
H_alpha_2=v0_2**2*np.sin(alpha_var)**2/2/g

v0_var=np.linspace(0,100,1000)
H_v0=v0_var**2*np.sin(alpha_rad)**2/2/g
H_v0_1=v0_var**2*np.sin(alpha_1_rad)**2/2/g
H_v0_2=v0_var**2*np.sin(alpha_2_rad)**2/2/g



fig,(ax0,ax1)=plt.subplots(1,2,figsize=(15,6))

plt.sca(ax0)
ax0.plot(alpha_var*180/np.pi,H_alpha,label=r'$v_0$='+str(v0)+'m/s')
ax0.plot(alpha_var*180/np.pi,H_alpha_1,label=r'$v_0$='+str(v0_1)+'m/s')
ax0.plot(alpha_var*180/np.pi,H_alpha_2,label=r'$v_0$='+str(v0_2)+'m/s')

ax0.set_xlabel(r'$\alpha$ (°)',fontsize=17)
ax0.set_ylabel('Flèche (m)', fontsize=17)
ax0.set_xlim(0,90)
ax0.set_ylim(0,120)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='upper right',fontsize=17)

plt.sca(ax1)
ax1.plot(v0_var,H_v0,label=r'$\alpha$='+str(alpha)+'°')
ax1.plot(v0_var,H_v0_1,label=r'$\alpha$='+str(alpha_1)+'°')
ax1.plot(v0_var,H_v0_2,label=r'$\alpha$='+str(alpha_2)+'°')
ax1.set_xlabel(r'$v_0$ (m/s)',fontsize=17)
ax1.set_ylabel('Flèche (m)', fontsize=17)
ax1.set_xlim(0,100)
ax1.set_ylim(0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=17)
plt.savefig('fleche')
```

# Chute avec frottements

```python
beta=1
m=1
alpha_drag=45
```

```python
tau_drag=m/beta
x_drag=v0*m/beta*(1-np.exp(-t/tau_drag))
y_drag=(v0*tau_drag*np.sin(alpha_drag)+g*tau_drag**2)*(1-np.exp(-t/tau_drag))-g*tau_drag*t
x_lim=v0*m/beta
```

```python
fig,ax=plt.subplots(figsize=(6,4))
ax.plot(x_drag,y_drag)
ax.plot([x_lim,x_lim],[7.5,-12.5],linestyle='--')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax.set_xlabel('x (m)',fontsize=15)
ax.set_ylabel('y (m)',fontsize=15)
```
