```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def s(t,T,a0=1,a1=1,a2=1,a3=1,a10=0):
    return(a0*np.sin(2*np.pi*t/T)+a1*np.sin(4*np.pi*t/T)+a2*np.sin(6*np.pi*t/T)+a3*np.sin(8*np.pi*t/T)+a10*np.sin(22*np.pi*t/T))
```

```python
t=np.linspace(0,10,1000)
T=0.002
a0_0,a1_0,a2_0,a3_0,a10_0=0.5,1,3,2,0
a0_1,a1_1,a2_1,a3_1,a10_1=0.5,1,0,0,0.1

sortie0=s(t,T,a0_0,a1_0,a2_0,a3_0,a10_0)
sortie1=s(t,T,a0_1,a1_1,a2_1,a3_1,a10_1)

fig,(ax0,ax1)=plt.subplots(1,2,figsize=(12,6))
ax0.plot(t,sortie0)
ax0.set_xlim(0,10)
plt.sca(ax0)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
ax0.set_xlabel('Temps (s)',fontsize=18)
ax0.set_ylabel('Amplitude',fontsize=18)

ax1.plot(t,sortie1)
plt.sca(ax1)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
ax1.set_xlabel('Temps (s)',fontsize=18)
#plt.savefig('signal_periodique.pdf')
```

```python
#### Spectre des deux signaux ####

f0=1/T
f1,f2,f3,f10=2*f0,3*f0,4*f0,11*f0
f=np.linspace(0,11*f0,10000)
a0=np.zeros(10000)
a1=np.zeros(10000)

f0_ind=np.where(abs(f-f0)<(11*f0/10000/2))[0][0]
f1_ind=np.where(abs(f-f1)<(11*f0/10000/2))[0][0]
f2_ind=np.where(abs(f-f2)<(11*f0/10000/2))[0][0]
f3_ind=np.where(abs(f-f3)<(11*f0/10000/2))[0][0]
f10_ind=np.where(abs(f-f10)<(11*f0/10000/2))[0][0]

harm_indices=[f0_ind,f1_ind,f2_ind,f3_ind,f10_ind]
print(f0_ind,f1_ind,f2_ind,f3_ind,f10_ind)
a0[harm_indices]=a0_0,a1_0,a2_0,a3_0,a10_0
a1[harm_indices]=a0_1,a1_1,a2_1,a3_1,a10_1


fig,(ax0,ax1)=plt.subplots(1,2,figsize=(12,4))
plt.subplots_adjust(bottom=0.2)
plt.sca(ax0)
ax0.plot(f/1000,a0)
ax0.set_xlabel('Fréquence (kHz)',fontsize=18)
ax0.set_ylabel('Amplitude',fontsize=18)
ax0.set_xlim(0)
ax0.set_ylim(0)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.sca(ax1)
ax1.plot(f/1000,a1)
ax1.set_xlabel('Fréquence (kHz)',fontsize=18)
ax1.set_xlim(0)
ax1.set_ylim(0)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

#plt.savefig('spectre.pdf')
```
