```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
tmax=10
t=np.linspace(0,tmax,1000)
tau=1
fig,ax=plt.subplots(figsize=(6,6))
ax.plot(t,np.exp(-t/tau),label='Second membre nul')
ax.plot(t,1-np.exp(-t/tau),label='Second membre non nul',linestyle='dashed')
ax.plot([0,tmax+tmax/10],[1,1],linestyle=':')
ax.set_xlim(0,tmax+tmax/10)
ax.set_ylim(0,1.1)
ax.set_xlabel('t (s)',fontsize=18)
plt.xticks(fontsize=16)
plt.yticks([0,1],['0','A'],fontsize=16)
plt.legend(fontsize=16)
#plt.savefig('effet_second_membre')
```

```python
tau_1,tau_2,tau_3=0.1,1,3
fig,ax=plt.subplots(figsize=(6,6))
ax.plot(t,1-np.exp(-t/tau_1),label=r'$\tau=$'+str(tau_1)+' s')
ax.plot(t,1-np.exp(-t/tau_2),label=r'$\tau=$'+str(tau_2)+' s',linestyle='dashed')
ax.plot(t,1-np.exp(-t/tau_3),label=r'$\tau=$'+str(tau_3)+' s',linestyle='--')
ax.plot([0,tmax+tmax/10],[1,1],linestyle=':')
ax.set_xlim(0,tmax+tmax/10)
ax.set_ylim(0,1.1)
ax.set_xlabel('t (s)',fontsize=18)
plt.xticks(fontsize=16)
plt.yticks([0,1],['0','A'],fontsize=16)
plt.legend(fontsize=16)
plt.savefig('effet_tau')
```
